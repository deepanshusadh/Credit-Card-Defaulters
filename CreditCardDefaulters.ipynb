```python
x=data.drop(labels=['default payment next month'],axis=1)
y=data['default payment next month']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_df=scaler.fit_transform(x)
```

```python
scale_df=pd.DataFrame(data=scaled_df, columns=x.columns, index=x.index)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(scale_df, y, test_size=0.33, random_state=42)
```

```python
x_test.shape
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(x_train)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from kneed import KneeLocator
kn = KneeLocator(range(1,11), wcss, curve='convex', direction='decreasing')
print(kn.knee)
```

```python
def naive_bayes(train,labels):
    gnb=GaussianNB()
    param_grid = {"var_smoothing": [1e-9,0.1, 0.001, 0.5,0.05,0.01,1e-8,1e-7,1e-6,1e-10,1e-11]}
#Creating an object of the Grid Search class
    grid = GridSearchCV(estimator=gnb, param_grid=param_grid, cv=4)
    grid.fit(train,labels)
    p=grid.best_params_['var_smoothing']
    gnb_best=GaussianNB(var_smoothing=p)
    gnb_best.fit(train,labels)
    return gnb_best
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
def Xg_boost(train,labels):
    param_grid_xgboost = {"n_estimators": [50,100, 130],"max_depth": [5,7,11]}
# Creating an object of the Grid Search class
    grid= GridSearchCV(XGBClassifier(objective='binary:logistic'),param_grid_xgboost, verbose=3,cv=4,n_jobs=-1)
    grid.fit(train,labels)
    n_esti=grid.best_params_['n_estimators']
    max_dep=grid.best_params_['max_depth']
    xgb_best=XGBClassifier(n_estimators=n_esti,max_depth=max_dep)
    xgb_best.fit(train,labels)
    return xgb_best
    
```
